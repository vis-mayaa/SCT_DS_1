```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
# Load the CSV file
file_path = '/content/unemployment analysis.csv'
data = pd.read_csv(file_path)
```

```python
# Remove missing values in 2021 column
data_2021 = data[['Country Name', '2021']].dropna()
```

```python
# Sort by 2021 values descending, pick top 10
top10 = data_2021.sort_values(by='2021', ascending=False).head(10)
```

```python
# Plot bar chart
plt.figure(figsize=(12,6))
sn.barplot(x='2021', y='Country Name', data=top10, palette='viridis')
plt.title('Top 10 Countries by Value in 2021')
plt.xlabel('Value')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```
